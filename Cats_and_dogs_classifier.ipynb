```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Convolution2D
from keras.layers import MaxPooling2D, MaxPool2D, BatchNormalization, Dropout, Flatten


from keras.layers import Dense
from keras.layers import Flatten


from tensorflow.keras.layers import Input, Conv2D, Activation
from tensorflow.keras.optimizers import Adam, Adadelta
import os
```

```python
#part2-fitting the cnn to the images
from keras.preprocessing.image import ImageDataGenerator
train_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)
```

```python
# Generating images for the Test set
test_datagen = ImageDataGenerator(rescale = 1./255)
```

### Train, test split

```python
from tensorflow.keras.utils import image_dataset_from_directory
from tensorflow.keras.preprocessing import image_dataset_from_directory

# Create datasets
train_datagen = image_dataset_from_directory(input_dir_img,
                                                  image_size=(128,128),
                                                  subset='training',
                                                  seed = 1,
                                                 validation_split=0.1,
                                                  batch_size= 32)
test_datagen = image_dataset_from_directory(input_dir_img,
                                                  image_size=(128,128),
                                                  subset='validation',
                                                  seed = 1,
                                                 validation_split=0.1,
                                                  batch_size= 32)
```

### Create model

```python
Image_Width=128
Image_Height=128
Image_Size=(Image_Width,Image_Height)
Image_Channels=3
```

```python
from keras.preprocessing.image import ImageDataGenerator,load_img
from keras.utils import to_categorical
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import random
import os

model=Sequential()
model.add(Conv2D(32,(3,3),activation='relu',input_shape=(Image_Width,Image_Height,Image_Channels)))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))
model.add(Conv2D(64,(3,3),activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))
model.add(Conv2D(128,(3,3),activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(512,activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.5))
model.add(Dense(2,activation='sigmoid'))
```

```python
# Model Compilation
model.compile(optimizer = Adam(learning_rate=0.001),
              loss = "sparse_categorical_crossentropy",
              metrics = "accuracy")
```

```python
model.summary()
```

### Apply the model to train dataset

```python
# Model Train

model.fit(train_datagen,epochs=10,validation_data=test_datagen)
```
